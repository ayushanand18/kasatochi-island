## grabbing data from GBIF
This is the biodiversity data (in case if needed).

```python
out = requests.get("https://www.gbif.org/api/occurrence/search?advanced=false&geometry=POLYGON((-175.67051+52.24139,-175.70031+52.16174,-175.63575+52.06642,-175.63475+52.06476,-175.47113+52.03063,-175.35844+52.12653,-175.35952+52.21972,-175.45271+52.29177,-175.58978+52.30586,-175.67051+52.24139))&has_coordinate=true&has_geospatial_issue=false&year=1995,2019&limit=1&offset=0")
res = out.json()["results"]
df = pd.DataFrame(columns = pd.DataFrame(res).columns)

for i in range(0,3527,300):
    out = requests.get(f"https://www.gbif.org/api/occurrence/search?advanced=false&geometry=POLYGON((-175.67051+52.24139,-175.70031+52.16174,-175.63575+52.06642,-175.63475+52.06476,-175.47113+52.03063,-175.35844+52.12653,-175.35952+52.21972,-175.45271+52.29177,-175.58978+52.30586,-175.67051+52.24139))&has_coordinate=true&has_geospatial_issue=false&year=1995,2019&limit=300&offset={i}")
    res = out.json()["results"]
    df = pd.concat([df,pd.DataFrame(res)])

df
```
